```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 3, 5])
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
plt.figure()
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Scatter Plot')
plt.show()
# Customized scatter plot
plt.figure()
plt.scatter(x, y, marker='x', color='red', s=100)  # Increased marker siz
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Scatter Plot')
plt.grid(True)
plt.show()
# Independent scatter plot
x_custom = np.array([1.5, 2.5, 3.2, 4, 4.8])
y_custom = np.array([2.2, 5.1, 3.8, 7.9, 9.5])
plt.figure()
plt.scatter(x_custom, y_custom, marker='o', color='blue', s=50)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Independent Scatter Plot')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
# Basic bar chart
plt.figure()
plt.bar(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Bar Chart')
plt.show()
# Grouped bar chart
y1 = np.array([5, 5, 9, 11, 15])
y2 = np.array([3, 5, 7, 9, 11])
plt.figure()
plt.bar(x - 0.2, y1, color='b', width=0.4, label='Bar 1')
plt.bar(x + 0.2, y2, color='g', width=0.4, label='Bar 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Grouped Bar Chart')
plt.legend()
plt.grid(axis='y')
plt.show()
# Independent Bar Chart
x_custom = np.array(['A', 'B', 'C', 'D', 'E'])
y_custom = np.array([3, 7, 5, 9, 2])
plt.figure()
plt.bar(x_custom, y_custom, color='purple', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Independent Bar Chart')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
sizes = np.array([20, 30, 40, 50, 500])
# Basic bubble chart
plt.figure()
plt.scatter(x, y, s=sizes)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Bubble Chart')
plt.show()
# Colored bubble chart
colors = ['red', 'green', 'blue', 'yellow', 'orange']
plt.figure()
plt.scatter(x, y, s=sizes, c=colors)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Colored Bubble Chart')
plt.grid(True)
plt.show()
#Independent Bubble Chart
sizes_custom = np.array([50, 150, 200, 300, 400])
colors_custom = ['teal', 'magenta', 'brown', 'gray', 'cyan']
plt.figure()
plt.scatter(x, y, s = sizes_custom, c = colors_custom)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Independent Bubble Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Basic pie chart
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [20, 30, 40, 10]
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Basic Pie Chart')
plt.show()

# Exploded pie chart
explode = (0, 0, 0.1, 0)
colors = ['red', 'green', 'blue', 'yellow']
plt.figure()
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')  # Corrected autopct format
plt.title('Exploded Pie Chart')
plt.show()

# Independent Pie Chart
labels_custom = ['A', 'B', 'C', 'D']
sizes_custom = [15, 25, 45, 15]
plt.figure()
plt.pie(sizes_custom, labels=labels_custom, autopct='%1.1f%%', startangle=90)  # Fixed `startangle` argument
plt.title('Independent Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])

plt.figure(figsize=(10, 6))

# Combined Plot
plt.bar(x - 0.2, y1, color='b', width=0.4, label='Bar 1')
plt.plot(x + 0.2, y2, marker='o', linestyle='-', color='r', label='Line 2')  # Fixed string closure

# Adding annotations
for i, txt in enumerate(y1):
    plt.annotate(txt, (x[i] - 0.2, y1[i] + 0.5), textcoords="offset points", xytext=(0, 10), ha='center')  # Fixed textcoords

for i, txt in enumerate(y2):
    plt.annotate(txt, (x[i] + 0.2, y2[i] + 0.5), textcoords="offset points", xytext=(0, 10), ha='center')  # Fixed textcoords

# Labels, title, and other settings
plt.xlabel('X-axis')
plt.ylabel('Values')
plt.title('Composite Bar and Line Plot with Annotations')
plt.legend()
plt.grid(True, axis='y')
plt.axhline(y=5, color='grey', linestyle='--', label='Threshold')
plt.text(2, 6, "Interesting Point", fontsize=10, style='italic')

plt.show()
```

```python
import matplotlib.pyplot as plt
import squarify
import numpy as np

# Sample data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [20, 30, 40, 10]
colors = ['red', 'green', 'blue', 'yellow']

# Treemap
plt.figure(figsize=(8, 6))
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=0.8)
plt.title('Treemap Example')
plt.axis('off')
plt.show()

# Independent Treemap
labels_custom = ['A', 'B', 'C', 'D', 'E']
sizes_custom = [10, 25, 15, 35, 20]
colors_custom = ['cyan', 'magenta', 'yellow', 'teal', 'gray']

plt.figure(figsize=(8, 6))
squarify.plot(sizes=sizes_custom, label=labels_custom, color=colors_custom, alpha=0.8)  # Fixed variable name and added alpha
plt.title('Independent Treemap')
plt.axis('off')
plt.show()
```
