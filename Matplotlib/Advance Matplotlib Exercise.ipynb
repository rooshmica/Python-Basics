```python
import csv
import matplotlib.pyplot as plt
x = []
y = []
with open('data4.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots) # Skip header row
    for row in plots:
        x.append(float(row[1])) # Use the 'Height' column as x
        y.append(float(row[2])) # Use the 'Weight' column as y
plt.scatter(x, y)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Scatter Plot from CSV')
plt.show() 
```

```python
x = []
y = []
with open('data4.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots) # Skip header row
    for row in plots:
        x.append(float(row[1])) # Use the 'Height' column as x
        y.append(float(row[2])) # Use the 'Weight' column as y
plt.bar(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Chart from CSV')
plt.show()
```

```python
import numpy as np
x = []
y = []
with open('data4.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots) # Skip header row
    for row in plots:
        x.append(float(row[1])) # Use the 'Height' column as x
        y.append(float(row[2])) # Use the 'Weight' column as y
plt.scatter(x, y, s=np.array(y)*2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bubble Chart from CSV')
plt.show()
```

```python
labels = []
sizes = []
with open('data4.csv', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    next(plots) # Skip header row
    for row in plots:
        labels.append(row[0])
        sizes.append(float(row[1]))
plt.pie(sizes, labels=labels)
plt.title('Pie Chart from CSV')
plt.show()
```

```python
import json

with open('data.json', 'r') as jsonfile:
    data = json.load(jsonfile)
# Extract x values as sizes and date as labels
labels = []
sizes = []
for item in data:
    labels.append(item['date'])
    sizes.append(item['x'])
# Create the pie chart
plt.pie(sizes, labels=labels)
plt.title('Pie Chart from JSON')
plt.show()
```

```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('data4.csv')
# Convert the date column to a datetime type if it's not already
df['date'] = pd.to_datetime(df['date'])
# Set the date column as the index of the DataFrame
df.set_index('date', inplace=True)
# Plot the x variable against the date index
plt.plot(df['Weight'])
# Add a title and labels for the x and y axes
plt.title("Plot of Weight vs Date")
plt.xlabel("Date")
plt.ylabel("Weight")
# Show the plot
plt.show()
```
